```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../../Recursos')
import toolbook as toolbook
```

```python
# Load the dataset
dataset = pd.read_csv('../../Datos/Transformaciones/heart_disease_dataset_no_num.csv')
```

```python
# Preprocess the data
X_train, X_test, y_train, y_test = toolbook.preprocess_data(dataset, 'target', 0.25, 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Create a Decision Tree classifier
dt_clf = DecisionTreeClassifier(random_state=50)
```

```python
# Train the Decision Tree classifier on the training data
dt_clf.fit(X_train, y_train)
```

```python
# Make predictions on the testing data using the Decision Tree classifier
y_pred_dt = dt_clf.predict(X_test)
```

```python
toolbook.evaluate_classification(y_pred_dt, y_test, "B")
```

```python
# Evaluate the performance of the Decision Tree classifier
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Decision Tree Accuracy:", accuracy_dt)
```

```python
# Now for the new dataset prediction
new_dataset = pd.read_csv('../../Datos/Prediciones/prueba_prediccion.csv')
y_new_dt = dt_clf.predict(new_dataset)
```

```python
# Prediction probabilities for the new dataset
y_new_prob_dt = dt_clf.predict_proba(new_dataset)
```

```python
# Extract the probabilities for class 0 and class 1
prob_absence_dt = y_new_prob_dt[0][0]
prob_presence_dt = y_new_prob_dt[0][1]
```

```python
# Calculate the percentages of absence and presence
percentage_absence_dt = prob_absence_dt * 100
percentage_presence_dt = prob_presence_dt * 100
```

```python
# Show the results
print("Percentage of absence of the medical condition (Decision Tree):", percentage_absence_dt, "%")
print("Percentage of presence of the medical condition (Decision Tree):", percentage_presence_dt, "%")
```
