```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import sys
sys.path.append('../../Recursos')
import toolbook as toolbook
```

```python
# Leer el dataset
dataset = pd.read_csv('../../Datos/Transformaciones/heart_disease_dataset_no_target.csv')
```

```python
# Preprocesar los datos
X_train, X_test, y_train, y_test = toolbook.preprocess_data(dataset, 'num', 0.25, 1)
```

```python
# Entrenar un clasificador SVM con los mismos parámetros que el árbol de decisiones
svm_classifier = SVC(kernel='rbf', C=1.0, gamma='scale')  
svm_classifier.fit(X_train, y_train)
```

```python
# Predecir en el conjunto de prueba usando SVM
y_pred_svm = svm_classifier.predict(X_test)
```

```python
# Evaluar el modelo SVM
toolbook.evaluate_classification(y_pred_svm, y_test, "M")
```

```python
# Leer el nuevo dataset
new_dataset = pd.read_csv('../../Datos/Prediciones/prueba_prediccion.csv')
```

```python
# Realizar la predicción usando SVM en el nuevo dataset
y_new_pred_svm = svm_classifier.predict(new_dataset)

# Mostrar los resultados para cada instancia del nuevo dataset usando SVM
for i, pred in enumerate(y_new_pred_svm):
    print("Predicción para la instancia", i+1, "usando SVM:", pred)
    print("Probabilidades para cada clase usando SVM:", y_new_pred_svm[i])
```
